# FBHCP Mitigation Summary

Current spatial output has one row for each type of incidental take activity with mitigation for each rzone for each species.

```python
import os
import geopandas as gpd
import numpy as np
```

```python
fullpath = '/Users/mflaxman/Downloads/isias_proj1_scen38_20171026/mitigation.shp'
mitigation = gpd.read_file(fullpath)
mitigation.head()
```

```python
# where did mitigation take place?
habitat_mitigation_actions = mitigation[mitigation.MITIGATION > 0]
habitat_mitigation_actions.head()
```

```python
mitigation_sum_by_species_report  = habitat_mitigation_actions.pivot_table(index=['BALANCE_TA','SPE_NAME'], values=['MITIGATION','MNG_MITIGA','ACRES'],
                               aggfunc=np.sum, margins=True, fill_value=0)
mitigation_sum_by_species_report
```

```python
mitigation_sum_by_species_report  = habitat_mitigation_actions.pivot_table(index=['BALANCE_TA','SPE_NAME'], values=['AGG_NAME'],
                               aggfunc=np.size, margins=True, fill_value=0)
mitigation_sum_by_species_report
```

```python
# where did management change take place?
management_mitigation_actions = mitigation[mitigation.MNG_MITIGA > 0]
management_mitigation_actions.head()
```
