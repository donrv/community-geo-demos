### Replace text to remove references to 'medium'

```python
impacts['activity'] = impacts['activity'].replace(to_replace='Medium ', value='', regex=True)
```

Filter results to impacts only, excluding Gopher Tortoise
(from full set including recommended mitigation activities)

```python
impacts = impacts[ (impacts.impact > 0) & (impacts.species != 'Gopher Tortoise')]
impacts
```

```python
droplist = {'id','SCENARIO','scenario','SC_NAME','SC_VERSION','SPE_CODE',\
                      'TIMESTEP','ACT_CODE','IMP_AREA','IMPACT_ID',\
                      'FIPSSTCO','MONUMENT_N','HAB_PATCH'\
                      'LORD', 'EASE', 'BUY', 'PRED', 'geometry',
                       'mitigation','management_mitigation','action_names',\
                       'unchanged'}
for d in droplist:
    if d in impacts.columns:
        impacts.drop(columns=[d],inplace=True)
```

```python
# make an ACT_GROUP for permit types
impacts['activity'].unique()
```

```python
def nonArmoringActivities(row):
    if 'Armoring' in row['activity']:
        return False
    else:
        return True
```

```python
impacts['Nonarmoring'] = impacts.apply(nonArmoringActivities, axis=1)
impacts['Nonarmoring'].unique()
```

```python
def groupActivities(row):
    val = ''
    if row['activity'] == 'Direct - Large Non-native Landscape, Pools & Decks':
        val = 'Coastal Construction - Landscape, Auxilliary Stuctures & Fill'

    elif row['activity'] == 'Direct - Small Non-native Landscape, Pools & Decks':
        val = 'Coastal Construction - Landscape, Auxilliary Stuctures & Fill'
    
    elif row['activity'] == 'Direct - Building Footprint - Small':
        val = 'Coastal Construction - SFR - Direct'
        
    elif row['activity'] == 'Direct - Building Footprint - Large':
        val = 'Coastal Construction - MFR/COM/GOV - Direct'
        
    elif row['activity'] == 'Indirect - Fill permits':
        val = 'Coastal Construction - Landscape, Auxilliary Stuctures & Fill'
        
    elif row['activity'] == 'Indirect - Multifamily/Commercial/Institutional Activities':
        val = 'Coastal Construction - MFR/COM/GOV - Indirect'
        
    elif row['activity'] == 'Indirect - Single Family Activity':
        val = 'Coastal Construction - SFR - Indirect'
        
    elif row['activity'] == 'xCommercial':
        val = 'Coastal Construction - MFR/COM/GOV - Indirect'

    elif row['activity'] == 'Direct - Dune Crossovers':
        val = 'Dune Crossovers - Direct'
    elif row['activity'] == 'Indirect - Dune Crossovers (SFR)':
        val = 'Dune Crossovers - Indirect'
    elif row['activity'] == 'Indirect - Dune Crossovers Public':
        val = 'Dune Crossovers - Indirect'
    else:
        val = 'Misc'
    return val
```

```python
impacts['ACT_GRP'] = impacts.apply(groupActivities, axis=1)
impacts['ACT_GRP'].unique()
```

```python
impacts['ACT_GRP'].unique()
```

```python
# Compute attribute of Directness
def isDirect(row):
    # shortcut
    d = row['activity'].split(' - ')[0]
    return(d)
```

```python
impacts['Directness'] = impacts.direct_impact
impacts['Directness'].head()
```

```python
impacts.species.unique()
```

```python
def groupSpecies(row):
    val = ''
    if row['species'] == 'Perdido Key beach mouse':
        val = 'Beach Mice'
    elif row['species'] == 'Saint Andrews Beach Mouse':
        val = 'Beach Mice'
    elif row['species'] == 'Choctawhatchee Beach mouse':
        val = 'Beach Mice'
    elif row['species'] == 'Santa Rosa Beach Mouse':
        val = 'Beach Mice'
    elif row['species'] == 'Anastasia Island Beach Mouse':
        val = 'Beach Mice'
    elif row['species'] == 'Southeastern beach mouse':
        val = 'Beach Mice'


    elif row['species'] == 'Solitary Nesting Shorebirds':
        val = 'Birds'
    elif row['species'] == 'Colonial Nesting Shorebirds':
        val = 'Birds'
    elif row['species'] == 'Nonbreeding Shorebirds':
        val = 'Birds'

    elif row['species'] == 'Seaturtles':
        val = 'Reptiles'
    elif row['species'] == 'Gopher Tortoise':
        val = 'Reptiles'

    return val
```

```python
impacts['SPE_GRP'] = impacts.apply(groupSpecies, axis=1)
impacts['SPE_GRP'].head()
```

```python
#impacts = impacts.reindex(['Nonbreeding Shorebirds', 'Solitary Nesting Shorebirds', 'Colonial Nesting Shorebirds'])
impacts.head(5)
```

```python
def Region(row):
    if row['balance_target'] == 'HCP Regions - NW':
        return('Panhandle')
    if row['balance_target'] == 'HCP Regions - NE':
        return('Northeast')
    if row['balance_target'] == 'HCP Regions - SW':
        return('Gulf')
    if row['balance_target'] == 'HCP Regions - SE':
        return('Southeast')
```

```python
impacts['FBHCP_Regn'] = impacts.apply(Region, axis=1)
```

### Need to extract R-zone name from aggregationarea

```python
def rzoneFromAgArea(row):
    aa = row['aggregationarea']
    ignore, rzone = aa.split(' - ')
    return rzone
```

```python
impacts['R-zone'] = impacts.apply(rzoneFromAgArea, axis=1)
impacts['R-zone'].unique()
```

```python
def direct2num(row):
    if (row['direct_impact'] is True):
        return(1)
    else:
        return(0)
    
impacts['DirectNo'] = impacts.apply(direct2num, axis=1)

impacts['Ac Direct'] = impacts['DirectNo'] * (impacts.acres)
impacts['Ac Indirect'] = (impacts['DirectNo']==0) * (impacts.acres)

impacts['QA Direct'] = impacts['DirectNo'] * impacts.acres * (impacts.quality)
impacts['QA Indirect'] = (impacts['DirectNo']==0) * impacts.acres * (impacts.quality)

impacts.head()
```
